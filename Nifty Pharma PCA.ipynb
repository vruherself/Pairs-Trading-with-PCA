```python
def get_return(close):
    r = close.pct_change(1)
    return r.dropna()
```

```python
def standardized_return(close):
    r = get_return(close)
    std_ret = (r - r.mean())/r.std()
    return std_ret.dropna()
```

```python
def get_num_components(df, var_ret):
    """
    Returns the number of components required to retain a particular variance(var_ret)
    """
    #variance retained should be in percentages between 0 and 1
    if var_ret > 1 or var_ret < 0:
        print('Error')
        return 0
    
    if var_ret == 1:
        return df.shape[1]
        
    pca = PCA(n_components = df.shape[1])

    pca.fit(df)
    
    needed_components = 0
    
    var_sum = 0
    
    for i in range(0, df.shape[1]):

        if var_sum >= var_ret:
            print('Total Variance Retained: ', pca.explained_variance_ratio_[0:needed_components].sum())
            return needed_components
        else:
            needed_components += 1            
            var_sum += pca.explained_variance_ratio_[i]
            
    
#num_components = get_num_components(returns.dropna(), 0.9)

#print('\nNumber of Principal Components Needed: ', num_components)
```

```python
def get_eigen(df, n_components):
    """Returns a dataframe of eigen vectors and eigenvalues with n components
    """
    pca = PCA(n_components)
    pca.fit(df)
    
    eigenvectors = pd.DataFrame(pca.components_, columns = df.columns)
    eigenvalues = pd.DataFrame(pca.singular_values_)
   
    return eigenvectors, eigenvalues
```

```python
def get_factor_ret(weights, returns):
    """
    Inputs :
    Weights = Eigenvector dataframe
    returns = non standarized returns
    
    We pass the eigenvector dataframe to get the corresponding eigenportfolio returns for the each eigenvector
    Output:
    Resultant dataframe will the the returns for the corresponding eigenvectors.
    The columns will represent the returns of the eigenvectors and the rows, the number of days under 
    consideration
    
    """
    if weights.shape[1] != weights.shape[1]:
        return 0,0
    #the weights for each stock is (v_i/std_i)
    weights_transform = weights/returns.std()
    
    
    factor_return = np.dot(returns, weights_transform.transpose())
    factor_return_matrix = pd.DataFrame(factor_return, index = returns.index)
    return factor_return_matrix
```

 So far we have the F matrix(which is the factor matrix) of the equation
 
 $$ r = \beta*F + \epsilon $$ 
 
 Now we will ahve to perform regression analysis for the $\beta$  part and get the residuals. Later model the residuals as the OU process
 
 So here,
 - F will have dimenasions (number of returns of days considered * # eigenvales considered)
 - r will have dimension of the number of days.
 - One thing to keep in mind here is that the while calculating the factor returns and while running the regression on the rest of the stocks to get the returns of the rest of the stocks, you need to train it on the data with same number of days

```python
def get_residual(returns , factor_return_matrix):
    #initialize the residual dataframe
    #here the returns can be of the other stocks that you might wanna trade
    
    res = pd.DataFrame(columns = returns.columns, index = returns.index)
    coef = pd.DataFrame(columns = returns.columns, index = range(16))
    ols = LinearRegression()
    
    for i in returns.columns:
        ols.fit(factor_return_matrix, returns[i])
        res[i] = returns[i]-ols.intercept_-np.dot(factor_return_matrix, ols.coef_)
        coef[i] = ols.coef_
    return res,coef
    
```
